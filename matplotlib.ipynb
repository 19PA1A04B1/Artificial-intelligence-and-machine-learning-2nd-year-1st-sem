# Plot styling

```
fig,ax=plt.subplots()

# First plot
x=np.arange(25)# x-axis is numbers from 0 to 49
y=np.sin(x)
ax.plot(x,y,color="green" ,linestyle="--",label="sin wave")

# Second plot
x1=np.arange(25)
y1=np.cos(x1)
ax.plot(x1,y1,color="red",linestyle=":",label="cos wave")

ax.legend(loc=4) 
# legend is a list of the names for each of the series on the chart.
# location of legend is an integer.
# for text to be correct in the legend the plots need a label


ax.set_xlabel("X axis")  # label of x-axis
ax.set_ylabel("y axis") # label of y-axis

ax.set_title("Sin and cos wave") # The heading of a graph
```
